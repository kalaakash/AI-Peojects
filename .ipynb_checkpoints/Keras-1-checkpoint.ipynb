#  Keras ML-project

```python
# Importing numpy sub-lib, Sequential model and Dense NN layers from Keras 
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Loading the datasets
dataset = loadtxt('pima-indian-diabetes.csv', delimiter=',')

# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
#Exploring the data set
print("First patients first attribute : " + str(X[0][0]))
print("Last patients last attribute : " + str(X[767][7]))
print("First patients output : " + str(y[0]))
print("Last patients output : " + str(y[767]))
```

```python
#Defining the model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150, batch_size=10)
```

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
predictions = model.predict_classes(X)
# summarize the first 5 cases
for i in range(5):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
